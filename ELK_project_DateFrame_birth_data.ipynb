# 데이터 전처리하기

row의 경우 연도별/ col의 경우 월별로 세팅하기

```python
mt = [i for i in range(1,13)]
yr = [i for i in range(1985,2021)]
```

```python
df1= pd.DataFrame(columns=yr, index=mt)
df1.head()
```

```python
i = 0
for year in range(1985, 2021):
    for month in range(1, 13):
        #df[행][열]
        df1[year][month] = df.iloc[month + i, 1]
    year += 1 # 열(년도)이 1개씩 커져야함
    i += 12 # df에서 12달씩 끊어야함    
df1
```

```python
df1.to_csv('df_birth_data.csv')
```
